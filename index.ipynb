```python
def show_distplot(dataframe, gender, column_name):
    plt.plot()
    dataframe[column_name].hist()
    plt.title(f'Distribution of {column_name} for {gender} heroes')
    plt.xlabel(column_name)
    plt.ylabel('Probability Density')
    plt.show()
```

### Sample Question: Most Common Powers

The rest of this notebook will be left to you to investigate the dataset by formulating your own questions, and then seeking answers using pandas and numpy. Before moving on to formulating your own questions, use the dataset to answer the following questions about superhero powers:

* What are the 5 most common powers overall?
* What are the 5 most common powers in the Marvel Universe?
* What are the 5 most common powers in the DC Universe?

```python
def top_5_powers(dataframe):
    # create a new dataframe without hero names
    df = dataframe.drop(heroes_df.columns.values[1:], axis=1)
    columns = df.columns.values
    
    # convert boolean values to integers
    for col in columns:
        df[col] = df[col].map({'True': 1, 'False': 0})
        
    # create a dictionary from column labels/ sums
    power_counts_dict = dict(df.sum())
    
    # sort the dictionary from highest-to-lowest count
    return sorted(power_counts_dict.items(), key=lambda x: x[1], reverse=True)[:5] 

# create a variable for the function's return of the top 5 powers
overall_top_5 = top_5_powers(heroes_and_powers_df)
# create a variable filtering the top 5 powers by the publisher `Marvel Comics`
marvel_df = heroes_and_powers_df[heroes_and_powers_df['Publisher'] == 'Marvel Comics']
# create a variable filtering the top 5 powers by the publisher `DC Comics`
dc_df = heroes_and_powers_df[heroes_and_powers_df['Publisher'] == 'DC Comics']

# print the overall most common powers
print(overall_top_5)
```

```python
# print the 5 most common powers for Marvel heroes

marvel_top_5 = top_5_powers(marvel_df)
print(marvel_top_5)
```

```python
# print the 5 most common powers for DC heroes

dc_top_5 = top_5_powers(dc_df)
print(dc_top_5)
```

```python
# make a dataframe of female heroes with the power of flight
female_heroes_flight_df = heroes_and_powers_df[(
    heroes_and_powers_df['Flight'] == 'True') & (
    heroes_and_powers_df['Gender'] == 'Female')].copy()

# view the shape and first 5 rows of the new dataframe
display(female_heroes_flight_df.shape)
female_heroes_flight_df.head()
```

```python
# find the top 5 powers that occur among female heroes with the power of flight
fem_flyer_top_5 = top_5_powers(female_heroes_flight_df)    

# print the overall most common powers
print(fem_flyer_top_5)
type(fem_flyer_top_5)
```

**What are the correlations among these powers?**

```python
# create a list of labels from `fem_flyer_top_5`
label_list = list(map(lambda k: k[0], fem_flyer_top_5))

# extract a new dataframe of female flyer top 5 powers
# using columns in label_list to filter `female_heroes_flight_df`
new_df = pd.DataFrame(female_heroes_flight_df, columns = label_list)

# delete the names column and reset the index
new_df.reset_index(drop=True, inplace=True)
    
# convert boolean values to integers
columns = new_df.columns.values
for col in columns:
    new_df[col] = new_df[col].map({'True': 1, 'False': 0})

# preview the new dataframe
new_df.head()
```

```python
# view correlations among top 5 powers of female heroes
# capable of flight
new_df.corr()
```

**The above method indicates that the powers most likely to co-occur among female
superheroes with the power of flight (excepting the power of flight itself) are
`Super Strength`/`SuperSpeed`, which have a correlation of 0.626947, followed by
`Super Strength`/`Durability`, which have a correlation of 0.485847.**

**An EDA of these correlations may be more clearly visualized via a Seaborn heatmap.**

```python
# perform visual EDA comparing correlations between the top 5 powers

# build a heatmapping function
def heatmapper(df):
    # create Correlation df
    corr = df.corr()
    # plot figsize
    fig, ax = plt.subplots(figsize=(8, 5))
    # generate heat map, allow annotations and place floats in map
    # `_r` to reverse colormap and make high correlations darker
    sns.heatmap(corr, cmap='viridis_r', annot=True, fmt=".3f")
    # show plot
    plt.show()
    
# plot new dataframe correlations through the heatmapping function
heatmapper(new_df)
```
